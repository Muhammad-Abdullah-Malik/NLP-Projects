```python
from transformers import AutoTokenizer
```

```python
tokenizer = AutoTokenizer.from_pretrained("bert-base-cased")
```

```python
def tokenize_and_align_labels(example):
    # tokenize each word independently
    tokenized = tokenizer(
        example["tokens"],
        is_split_into_words=True,
        truncation=True,
        padding="max_length",
        max_length=128
    )
    
    # map each subword to the original word index
    word_ids = tokenized.word_ids()
    
    label_ids = []
    previous_word_id = None
    
    for word_id in word_ids:
        
        # special tokens: CLS, SEP, PAD → ignore
        if word_id is None:
            label_ids.append(-100)
        
        # first subword of a word → assign label
        elif word_id != previous_word_id:
            label = example["labels"][word_id]
            label_ids.append(label2id[label])
        
        # other subwords → ignore
        else:
            label_ids.append(-100)
        
        previous_word_id = word_id

    tokenized["labels"] = label_ids
    return tokenized
```

```python
from datasets import Dataset

dataset = Dataset.from_list(sentences)
```

```python
dataset
```

```python
tokenized_dataset = dataset.map(tokenize_and_align_labels, batched=False)
```

```python
tokenized_dataset = tokenized_dataset.train_test_split(test_size=0.1)
train_dataset = tokenized_dataset["train"]
eval_dataset = tokenized_dataset["test"]
```

```python
from transformers import AutoModelForTokenClassification

model = AutoModelForTokenClassification.from_pretrained(
    "bert-base-cased",
    num_labels=len(label2id),
    id2label=id2label,
    label2id=label2id
)
```

```python
from transformers import TrainingArguments

training_args = TrainingArguments(
    output_dir="./ingredient-ner-model",
    eval_strategy="epoch",
    save_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=5,
    weight_decay=0.01,
    logging_steps=50,
    report_to="none"
)
```

```python
from transformers import DataCollatorForTokenClassification

data_collator = DataCollatorForTokenClassification(tokenizer)
```

```python
import numpy as np
from seqeval.metrics import precision_score, recall_score, f1_score, accuracy_score

def compute_metrics(p):
    predictions, labels = p
    predictions = np.argmax(predictions, axis=2)
    
    true_predictions = []
    true_labels = []
    
    for pred, lab in zip(predictions, labels):
        temp_pred = []
        temp_lab = []
        
        for p_id, l_id in zip(pred, lab):
            if l_id != -100:
                temp_pred.append(id2label[p_id])
                temp_lab.append(id2label[l_id])
        
        true_predictions.append(temp_pred)
        true_labels.append(temp_lab)
    
    return {
        "precision": precision_score(true_labels, true_predictions),
        "recall": recall_score(true_labels, true_predictions),
        "f1": f1_score(true_labels, true_predictions),
        "accuracy": accuracy_score(true_labels, true_predictions)
    }
```

```python
from transformers import Trainer

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=eval_dataset,
    data_collator=data_collator,
    tokenizer=tokenizer,
    compute_metrics=compute_metrics
)

trainer.train()
```

```python
from transformers import AutoTokenizer, AutoModelForTokenClassification
import torch

model_path = "/kaggle/working/ingredient-ner-model/checkpoint-930"

tokenizer = AutoTokenizer.from_pretrained(model_path)
model = AutoModelForTokenClassification.from_pretrained(model_path)
model.eval()
```

```python
def predict_ingredients(text):
    # Tokenize
    tokens = tokenizer(text.split(), is_split_into_words=True, return_tensors="pt")
    
    # Predict
    with torch.no_grad():
        outputs = model(**tokens)
    
    # Get predictions per token
    logits = outputs.logits
    predictions = torch.argmax(logits, dim=2)
    
    # Convert IDs -> labels
    pred_labels = [id2label[p.item()] for p in predictions[0]]
    
    # Re-align tokens (remove padding + special tokens)
    word_ids = tokens.word_ids()
    
    final_tokens = []
    final_labels = []
    seen_words = set()
    
    for token, label, word_id in zip(tokens.tokens(), pred_labels, word_ids):
        if word_id is None:  
            continue
        
        if word_id not in seen_words:  # first subword only
            final_tokens.append(token)
            final_labels.append(label)
            seen_words.add(word_id)
    
    return list(zip(final_tokens, final_labels))
```
